````python
import base64
import time
import re
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords

# Download NLTK stopwords if not already present
nltk.download('stopwords', quiet=True)

def get_readme(repo_url, headers=None):
    """Fetches README content from GitHub API"""
    # Extract owner and repo from github URL
    _, _, _, owner, repo = repo_url.rstrip('/').split('/')
    
    # List of possible README filenames
    readme_names = ['README.md', 'Readme.md', 'readme.md', 'README.markdown']
    
    headers = headers or {'Accept': 'application/vnd.github.v3+json'}
    
    for name in readme_names:
        try:
            url = f'https://api.github.com/repos/{owner}/{repo}/contents/{name}'
            response = requests.get(url, headers=headers)
            
            if response.status_code == 200:
                content = base64.b64decode(response.json()['content']).decode('utf-8')
                return content
                
        except Exception as e:
            continue
    
    return None

def preprocess_readme(raw_readme):
    """
    Preprocesses a README string:
    - Removes HTML tags and markdown images/logos
    - Removes code blocks and inline code
    - Removes URLs
    - Removes non-alphanumeric characters (except spaces)
    - Converts to lowercase
    - Removes stopwords
    - Returns cleaned text
    """
    if raw_readme is None:
        return ""
    
    # Remove HTML tags if any
    soup = BeautifulSoup(raw_readme, "html.parser")
    text = soup.get_text()
    
    # Remove markdown images: ![alt](url)
    text = re.sub(r'!\[.*?\]\(.*?\)', ' ', text)
    # Remove HTML <img> tags
    text = re.sub(r'<img[^>]*>', ' ', text)
    # Remove markdown links but keep the text: [text](url) -> text
    text = re.sub(r'\[([^\]]+)\]\([^\)]+\)', r'\1', text)
    # Remove code blocks (```...```)
    text = re.sub(r'```.*?```', ' ', text, flags=re.DOTALL)
    # Remove inline code (`...`)
    text = re.sub(r'`[^`]+`', ' ', text)
    # Remove URLs
    text = re.sub(r'http\S+|www\.\S+', ' ', text)
    # Remove non-alphanumeric characters (except spaces)
    text = re.sub(r'[^a-zA-Z0-9 ]', ' ', text)
    # Convert to lowercase
    text = text.lower()
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in text.split() if word not in stop_words]
    cleaned_text = ' '.join(tokens)
    return cleaned_text
````

````python
from typing import List
import google.generativeai as genai
import numpy as np

def summarize_reasoning_texts_with_gemini(reasonings: List[str], multipliers: List[float], api_key: str) -> str:
    """
    Summarizes a list of reasoning texts using the Gemini model.
    Reasonings with exceptionally high multipliers are given extra weight in the prompt.
    """
    # Compute high multiplier threshold (mean + 3*std)
    multipliers_arr = np.array([m if m is not None else 0 for m in multipliers])
    high_mult_threshold = multipliers_arr.mean() + 3 * multipliers_arr.std()

    # Separate high-multiplier and regular reasonings
    high_mult_reasonings = [
        r for r, m in zip(reasonings, multipliers)
        if isinstance(r, str) and r.strip() and m is not None and m > high_mult_threshold
    ]
    regular_reasonings = [
        r for r, m in zip(reasonings, multipliers)
        if isinstance(r, str) and r.strip() and (m is None or m <= high_mult_threshold)
    ]

    # Compose the prompt, giving extra emphasis to high-multiplier reasonings
    prompt = (
        "Summarize the following Ethereum ecosystem project comparison reasonings. "
        "Preserve all nuanced information, including minority opinions and subtle details, "
        "so that no underlying semantics are lost. "
        "Pay special attention to the reasonings that are associated with exceptionally high multiplier values, "
        "as these represent judgments of extraordinary impact or importance. "
        "The summary will be used as a prompt for a juror model.\n\n"
    )
    if high_mult_reasonings:
        prompt += (
            "The following reasonings are associated with very high multiplier values and should be given extra weight in your summary:\n"
            + "\n\n".join(high_mult_reasonings)
            + "\n\n"
        )
    prompt += (
        "The remaining reasonings are as follows:\n"
        + "\n\n".join(regular_reasonings)
    )

    # Initialize Gemini API
    genai.configure(api_key=api_key)
    model = genai.GenerativeModel('models/gemini-2.5-pro')

    # Generate summary
    response = model.generate_content(prompt)
    summary = response.text if hasattr(response, "text") else str(response)
    return summary

all_reasoning = train_latest['reasoning'].tolist()
all_multipliers = train_latest['multiplier'].tolist()

# You must provide your Gemini API key here
# GEMINI_API_KEY = ""
GEMINI_API_KEY = "<REPLACE_WITH_YOUR_GEMINI_API_KEY>"

# Actually get the summary from Gemini, giving extra weight to high-multiplier reasonings
latest_reasoning_summary = summarize_reasoning_texts_with_gemini(all_reasoning, all_multipliers, GEMINI_API_KEY)

# The variable `reasoning_summary` now contains the summary to be used as a prompt.
latest_reasoning_summary
````

````python
import json
import re
import csv
import time
import os
import google.generativeai as genai

# --- CONFIGURATION ---
LLM_MODEL = "models/gemini-2.5-pro"
API_KEY = "<REPLACE_WITH_YOUR_GEMINI_API_KEY>"
API_SLEEP_SECONDS = 15  # Be gentle: wait this many seconds between LLM calls

# --- DATA (Manually collected data) ---
client_market_data = {
    'geth': 61.72, 'nethermind': 18.57, 'erigon': 7.17, 'besu': 6.44, 
    'reth': 5.56, 'lighthouse': 44.89, 'prysm': 24.68, 'teku': 11.98, 
    'nimbus': 10.01, 'lodestar': 2.18, 'grandine': 0.28
}

package_downloads = {
    "ethers-io/ethers.js": 1750502, "nomicfoundation/hardhat": 74467,
    "openzeppelin/openzeppelin-contracts": 463048, "wevm/viem": 1787316,
    "ethereum/remix-project": 687, "apeworx/ape": 14234,
    "ethereum/web3.py": 714558, "vyperlang/titanoboa": 495,
    "alloy-rs/alloy": 76000
}

# --- LLM PROMPT DEFINITIONS ---
PHASE_1_PROMPT = """
You are an expert analyst tasked with creating a funding allocation model for the Ethereum ecosystem. Your goal is to distribute 100 total "contribution points" among several categories of open-source projects.

Your allocation must be guided by the established reasoning of expert human jurors. A summary of their collective reasoning is as follows:

**Juror Reasoning Summary:**
- **Primacy of the Core Protocol:** Jurors have a clear hierarchy. They consistently value core protocol infrastructure (Execution and Consensus Clients) that is essential for the network's survival and security far more than supplementary developer tools or libraries.
- **Real-World Adoption and Impact:** The most heavily weighted factor is a project's proven, current impact, measured by hard data like market share and download counts. The "disappearance test" (how badly would the ecosystem be affected if this project vanished?) is a key mental model.
- **Ecosystem Health & Decentralization:** Contributions that strengthen the entire network are highly valued. Specifically, projects that support client diversity are considered critical for network resilience.
- **Developer Enablement (Force Multiplier):** Foundational tools and libraries that empower a vast number of other developers to build are seen as having a massive downstream impact.

**TASK:**
Based on the juror reasoning summary above, distribute a total of 100 points among the following project categories. The points should reflect each category's overall importance to the Ethereum universe.

**Project Categories:**
{categories}

Provide your output as a simple list in the format `Category Name: Points`. The total must sum to 100.
"""

PHASE_2_PROMPT_TEMPLATE = """
You are an expert analyst tasked with scoring the relative contribution of several open-source projects within the same category.

Your analysis must be guided by the established reasoning of expert human jurors. A summary of their collective reasoning is as follows:

**Juror Reasoning Summary:**
- **Primacy of the Core Protocol:** Core infrastructure is valued most.
- **Real-World Adoption and Impact:** Value is measured by hard data like market share and download counts.
- **Ecosystem Health & Decentralization:** Client diversity is critical.
- **Developer Enablement (Force Multiplier):** Empowering other developers is a massive plus.
- **Technical Innovation & Maturity:** Unique, "battle-tested," and well-maintained projects are preferred.

**TASK:**
You are evaluating the "{category_name}" category. For each repository below, assign a relative score from 1 to 100, where 100 represents the highest possible impact *within this group*. Provide a brief, one-sentence rationale for your score, referencing the juror criteria and the provided data.

**Repository Data:**
{repo_data_json}

Provide your output as a valid JSON list of objects in the following format. Do not include any text outside of the JSON list itself.
[
  {{
    "repo": "owner/repo-A",
    "score": YOUR_SCORE,
    "rationale": "Your brief justification here, based on the data and juror principles."
  }}
]
"""

def call_llm(prompt):
    """
    Calls the Gemini 2.5 Pro model with a given prompt using the real API key.
    Returns the model's response text.
    """
    genai.configure(api_key=API_KEY)
    model = genai.GenerativeModel(LLM_MODEL)
    try:
        response = model.generate_content(prompt)
        # The Gemini API returns a response object with a .text attribute
        if hasattr(response, "text"):
            return response.text
        elif hasattr(response, "candidates") and response.candidates:
            # Fallback for some SDK versions
            return response.candidates[0].content.parts[0].text
        else:
            print("WARNING: Unexpected Gemini response format.")
            return str(response)
    except Exception as e:
        print(f"ERROR: Gemini API call failed: {e}")
        return None

def correct_llm_json_response(response_text):
    """
    Attempts to correct common LLM JSON formatting errors.
    """
    cleaned = response_text.strip()
    cleaned = re.sub(r"^```json", "", cleaned)
    cleaned = re.sub(r"^```", "", cleaned)
    cleaned = re.sub(r"```$", "", cleaned)
    cleaned = cleaned.strip()
    cleaned = re.sub(r",\s*]", "]", cleaned)
    cleaned = re.sub(r",\s*}", "}", cleaned)
    if cleaned.startswith("{") and cleaned.endswith("}"):
        cleaned = "[" + cleaned + "]"
    return cleaned

def load_existing_scores(csv_filename):
    """
    Loads existing repo scores from a CSV file, if it exists.
    Returns a dict: {repo: score}
    """
    scores = {}
    if os.path.exists(csv_filename):
        try:
            with open(csv_filename, mode='r', encoding='utf-8') as csvfile:
                reader = csv.DictReader(csvfile)
                for row in reader:
                    repo = row.get('repo')
                    score = row.get('score')
                    if repo and score:
                        try:
                            scores[repo] = int(score)
                        except Exception:
                            continue
        except Exception as e:
            print(f"WARNING: Could not read existing scores from {csv_filename}: {e}")
    return scores

def save_scores_to_csv(csv_filename, repo_scores):
    """
    Writes repo_scores (dict) to CSV, overwriting or creating the file.
    """
    try:
        with open(csv_filename, mode='w', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(['repo', 'score'])
            for repo, score in repo_scores.items():
                writer.writerow([repo, score])
        print(f"\nCSV file '{csv_filename}' has been updated with columns: repo, score.")
    except Exception as e:
        print(f"ERROR: Could not write CSV file: {e}")

def main():
    # 1. Load all prerequisite data files
    print("Loading data files...")
    try:
        # results is assumed to be a list of dicts with repo data, already available in memory
        with open('repo_categories.json', 'r') as f:
            repo_configs = json.load(f)
    except FileNotFoundError as e:
        print(f"FATAL ERROR: Make sure '{e.filename}' exists. Exiting.")
        return

    # 2. Enrich results with category and any extra data
    print("Enriching repo data with categories and metrics...")
    enriched_results = []
    for repo_info in results:
        repo_url = repo_info.get("repo")
        if repo_url and "github.com/" in repo_url:
            repo_slug = "/".join(repo_url.split('/')[-2:])
        else:
            repo_slug = repo_url
        repo_info['repo'] = repo_slug
        config = repo_configs.get(repo_slug, {})
        repo_info['category'] = config.get("category", "General")
        # Attach market share if relevant
        if config.get("category") in ["Execution Client", "Consensus Client", "Light Client"]:
            lookup_name = config.get("lookup_name")
            if lookup_name in client_market_data:
                repo_info['market_share'] = client_market_data[lookup_name]
        # Attach weekly downloads if available
        if repo_slug in package_downloads:
            repo_info['weekly_downloads'] = package_downloads[repo_slug]
        enriched_results.append(repo_info)

    # 3. Get all unique categories
    categories = sorted(list(set(repo['category'] for repo in enriched_results)))

    # --- PHASE 1: Get Inter-Category Weights ---
    print("\n--- STARTING PHASE 1: CATEGORY WEIGHTING ---")
    phase1_formatted_prompt = PHASE_1_PROMPT.format(categories='\n'.join(f"- {c}" for c in categories))
    phase1_response = call_llm(phase1_formatted_prompt)
    if not phase1_response:
        print("FATAL ERROR: Could not get response for Phase 1. Exiting.")
        return

    # Parse category weights
    category_weights = {}
    for line in phase1_response.strip().split('\n'):
        match = re.match(r".*?([\w\s&/]+?):\s*(\d+)", line.strip())
        if match:
            category_weights[match.group(1).strip()] = int(match.group(2))
    print("\n>>> Gemini 2.5 Pro Response for Category Weights:")
    print(json.dumps(category_weights, indent=2))

    # --- PHASE 2: Get Intra-Category Scores ---
    print("\n--- STARTING PHASE 2: REPOSITORY SCORING ---")
    csv_filename = "repo_scores.csv"
    repo_scores = load_existing_scores(csv_filename)
    print(f"Loaded {len(repo_scores)} existing repo scores from {csv_filename}.")

    repos_by_category = {cat: [] for cat in categories}
    for repo in enriched_results:
        repos_by_category[repo['category']].append(repo)

    for category, repos in repos_by_category.items():
        if not repos:
            continue
        print(f"\n--- Processing category: {category} ---")
        # Only process repos in this category that are not already scored
        unscored_repos = [repo for repo in repos if repo['repo'] not in repo_scores]
        if not unscored_repos:
            print(f"  - All repos in category '{category}' already scored. Skipping.")
            continue

        repo_dossiers = []
        for repo in unscored_repos:
            dossier = {"repo": repo["repo"], "description": repo.get("description", "")}
            if 'stars' in repo: dossier['stars'] = repo['stars']
            if 'market_share' in repo: dossier['market_share'] = repo['market_share']
            if 'weekly_downloads' in repo: dossier['weekly_downloads'] = repo['weekly_downloads']
            repo_dossiers.append(dossier)

        phase2_formatted_prompt = PHASE_2_PROMPT_TEMPLATE.format(
            category_name=category,
            repo_data_json=json.dumps(repo_dossiers, indent=2)
        )
        phase2_response_text = call_llm(phase2_formatted_prompt)
        if not phase2_response_text:
            print(f"  - ERROR: Failed to get LLM response for category {category}.")
            continue

        try:
            cleaned_response = correct_llm_json_response(phase2_response_text)
            scores_list = json.loads(cleaned_response)
            print(f"\n>>> Gemini 2.5 Pro Response for {category}:")
            print(json.dumps(scores_list, indent=2))
            for score_item in scores_list:
                repo_scores[score_item['repo']] = score_item['score']
            # Save progress after each category
            save_scores_to_csv(csv_filename, repo_scores)
        except Exception as e:
            print(f"  - ERROR: Could not parse LLM response for category {category}.")
            print("Raw Response:", phase2_response_text)
            print("Exception:", e)
        # Be gentle with the API: sleep after each category
        print(f"Sleeping {API_SLEEP_SECONDS} seconds to be gentle with the API...")
        time.sleep(API_SLEEP_SECONDS)

    print("\n\n--- ALL PHASES COMPLETE ---")
    print("Review the outputs above to confirm the model's reasoning.")

    # Final save (in case nothing new was added, this is a no-op)
    save_scores_to_csv(csv_filename, repo_scores)

# This block allows you to run the script from the command line or a notebook cell
if __name__ == "__main__":
    main()
````
